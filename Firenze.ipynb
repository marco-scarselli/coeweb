```python
def da_html(cartella, output):
    elenco = {}

    for item in os.listdir(cartella):
        dato =  pd.read_html(os.path.join(cartella ,item),header= 0, index_col=0, encoding = "cp1252", infer_types=False)[1]
        #dato["verso"] =  item
        elenco[item] = dato

        print item, elenco[item].shape

    tutte_tabelle = pd.concat(elenco.values())
    tutte_tabelle = tutte_tabelle.stack().reset_index().drop_duplicates()
    tutte_tabelle[0] = tutte_tabelle[0].apply(convert_string_euro)# .(index = "Divisioni", columns = "level_1")
    tutte_tabelle = tutte_tabelle.pivot_table(index = "Divisioni", columns = "level_1", values = 0 )#.to_excel(os.path.join("xlsx", output + ".xlsx"))
    return tutte_tabelle
```

```python
def solo_colonne_exp(dataset):
    elenco = list(dataset.columns)
    nuovo_elenco = list()
    for item in elenco:
        if item[:3] == "EXP":
            nuovo_elenco.append(item)
    return dataset[nuovo_elenco]

def numeroindice(dataset, anno0):
    df_numero_indice = dataset.div(dataset[anno0], axis=0) *100
    return df_numero_indice

def numero_indice__exp(dataset, anno0):
    risultato = solo_colonne_exp(numeroindice(dataset, anno0))
    return risultato
```
